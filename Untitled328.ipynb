```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Data from the table
data = {
    'Model': ['ResNet 50', 'Vgg19', 'MobileNet V2'],
    'Weighted FedProx': [64.8, 69.2, 65.8],
    'Unweighted FedProx': [64.8, 69.2, 65.8]
}

# Convert data to pandas DataFrame
df = pd.DataFrame(data)

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.drop(columns='Model'), annot=True, cmap='YlGnBu', linewidths=.5)
plt.title('Accuracy using FedProx')
plt.ylabel('Model')
plt.xlabel('Algorithm')
plt.xticks(ticks=[0.5, 1.5], labels=['Weighted FedProx', 'Unweighted FedProx'])
plt.tight_layout()
plt.show()

# Stacked bar plot
plt.figure(figsize=(10, 6))
df.set_index('Model').plot(kind='bar', stacked=True, rot=0)
plt.title('Accuracy using FedProx')
plt.ylabel('Accuracy (%)')
plt.xlabel('Model')
plt.ylim(0, 200)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Hexbin plot
plt.figure(figsize=(8, 6))
plt.hexbin(df['Weighted FedProx'], df['Unweighted FedProx'], gridsize=10, cmap='Blues')
plt.colorbar(label='Count')
plt.title('Hexbin Plot: Weighted vs. Unweighted FedProx')
plt.xlabel('Weighted FedProx Accuracy (%)')
plt.ylabel('Unweighted FedProx Accuracy (%)')
plt.tight_layout()
plt.show()

# Dual axis plot
fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:red'
ax1.set_xlabel('Model')
ax1.set_ylabel('Weighted FedProx', color=color)
ax1.bar(df['Model'], df['Weighted FedProx'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Unweighted FedProx', color=color)
ax2.plot(df['Model'], df['Unweighted FedProx'], color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title('Dual Axis Plot: Weighted vs. Unweighted FedProx')
plt.show()

# 3D scatter plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

x = df['Weighted FedProx']
y = df['Unweighted FedProx']
z = range(len(df))

ax.scatter(x, y, z, c=z, cmap='Blues', s=100)
ax.set_xlabel('Weighted FedProx Accuracy (%)')
ax.set_ylabel('Unweighted FedProx Accuracy (%)')
ax.set_zlabel('Model Index')
plt.title('3D Scatter Plot: Weighted vs. Unweighted FedProx')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data from the table
data = {
    'Model': ['ResNet 50', 'Vgg19', 'MobileNet V2'],
    'Weighted FedProx': [64.8, 69.2, 65.8],
    'Unweighted FedProx': [64.8, 69.2, 65.8]
}

# Convert data to pandas DataFrame
df = pd.DataFrame(data)

# Line plot with markers
plt.figure(figsize=(10, 6))
for algo in ['Weighted FedProx', 'Unweighted FedProx']:
    plt.plot(df['Model'], df[algo], marker='o', label=algo)
plt.title('Accuracy using FedProx')
plt.ylabel('Accuracy (%)')
plt.xlabel('Model')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.legend(title='Algorithm', loc='lower right')
plt.tight_layout()
plt.show()

# Step plot
plt.figure(figsize=(10, 6))
for algo in ['Weighted FedProx', 'Unweighted FedProx']:
    plt.step(df['Model'], df[algo], marker='o', label=algo)
plt.title('Accuracy using FedProx')
plt.ylabel('Accuracy (%)')
plt.xlabel('Model')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.legend(title='Algorithm', loc='lower right')
plt.tight_layout()
plt.show()

# Stem plot
plt.figure(figsize=(10, 6))
for algo in ['Weighted FedProx', 'Unweighted FedProx']:
    plt.stem(df['Model'], df[algo], label=algo, basefmt=' ', use_line_collection=True)
plt.title('Accuracy using FedProx')
plt.ylabel('Accuracy (%)')
plt.xlabel('Model')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.legend(title='Algorithm', loc='lower right')
plt.tight_layout()
plt.show()

# Spaghetti plot
plt.figure(figsize=(10, 6))
for algo in ['Weighted FedProx', 'Unweighted FedProx']:
    plt.plot(df['Model'], df[algo], label=algo)
plt.title('Accuracy using FedProx')
plt.ylabel('Accuracy (%)')
plt.xlabel('Model')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.legend(title='Algorithm', loc='lower right')
plt.tight_layout()
plt.show()

# Lollipop plot
plt.figure(figsize=(10, 6))
for algo in ['Weighted FedProx', 'Unweighted FedProx']:
    plt.stem(df['Model'], df[algo], linefmt='-', markerfmt='o', basefmt=' ', label=algo, use_line_collection=True)
plt.title('Accuracy using FedProx')
plt.ylabel('Accuracy (%)')
plt.xlabel('Model')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.legend(title='Algorithm', loc='lower right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Data from the table
data = {
    'Model': ['ResNet 50', 'Vgg19', 'MobileNet V2'],
    'Weighted FedProx': [64.8, 69.2, 65.8],
    'Unweighted FedProx': [64.8, 69.2, 65.8]
}

# Convert data to pandas DataFrame
df = pd.DataFrame(data)

# Radar chart
labels=np.array(['Weighted FedProx', 'Unweighted FedProx'])
stats=df.loc[0,labels].values
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
stats=np.concatenate((stats,[stats[0]]))
angles=np.concatenate((angles,[angles[0]]))
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, stats, color='blue', alpha=0.25)
ax.plot(angles, stats, color='blue', linewidth=2)
ax.set_yticklabels([])
plt.title('Radar Chart: Accuracy using FedProx')
plt.show()

# Parallel coordinates plot
plt.figure(figsize=(10, 6))
pd.plotting.parallel_coordinates(df, 'Model', color=('blue', 'green'))
plt.title('Parallel Coordinates Plot: Accuracy using FedProx')
plt.ylabel('Accuracy (%)')
plt.xlabel('Model')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Andrews curves plot
plt.figure(figsize=(10, 6))
pd.plotting.andrews_curves(df, 'Model', color=('blue', 'green'))
plt.title('Andrews Curves Plot: Accuracy using FedProx')
plt.xlabel('Model')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Pairplot
plt.figure(figsize=(8, 6))
sns.pairplot(df, kind='scatter', diag_kind='kde', markers='o')
plt.suptitle('Pairplot: Accuracy using FedProx', y=1.02)
plt.tight_layout()
plt.show()

# Jointplot with regression line
plt.figure(figsize=(8, 6))
sns.jointplot(x='Weighted FedProx', y='Unweighted FedProx', data=df, kind='reg')
plt.suptitle('Jointplot: Accuracy using FedProx', y=1.02)
plt.tight_layout()
plt.show()
```
